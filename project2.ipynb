```python
# Connect to the Cassandra Cluster from python
from cassandra.cluster import Cluster
cluster = Cluster(['localhost'], port=9042)  # replace with your Cassandra node IP address
session = cluster.connect()
```

```python
# Set up a keyspace
session.execute("Create KEYSPACE IF NOT EXISTS steamlit_app WITH REPLICATION = { 'class' : 'SimpleStrategy', 'replication_factor' : 1 };")
```

```python
#Create a table
session.set_keyspace('steamlit_app')
session.execute("DROP TABLE IF EXISTS user_data;")
session.execute("""
    CREATE TABLE IF NOT EXISTS user_data (
        user_id UUID PRIMARY KEY,
        name TEXT,
        age INT,
        email TEXT
    );
""")
```

```python
# Insert some data (ind is the primary key, must be unique)
session.execute("INSERT INTO user_data (user_id, name, age, email) VALUES (uuid(), 'Alice', 30, 'alice@example.com');")
session.execute("INSERT INTO user_data (user_id, name, age, email) VALUES (uuid(), 'Bob', 25, 'bob@example.com');")
session.execute("INSERT INTO user_data (user_id, name, age, email) VALUES (uuid(), 'Charlie', 35, 'charlie@example.com');")
```

```python
#Query the data
rows = session.execute("SELECT * FROM user_data;")
for row in rows:
    print(row)
    
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("SparkCassandraApp") \
    .config("spark.jars.packages", "com.datastax.spark:spark-cassandra-connector_2.12:3.1.0") \
    .config("spark.cassandra.connection.host", "127.0.0.1") \
    .config("spark.sql.extensions", "com.datastax.spark.connector.CassandraSparkExtensions") \
    .config("spark.sql.catalog.mycatalog", "com.datastax.spark.connector.datasource.CassandraCatalog") \
    .config("spark.cassandra.connection.port", "9042").getOrCreate()


```
